```python
import pandas as pd
df = pd.read_csv("Telco-Customer-Churn.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df['totalcharges'] = pd.to_numeric(
    df['totalcharges'],
    errors='coerce'
) 
```

```python
df['totalcharges'].fillna(0, inplace=True)
```

```python
df.duplicated().sum()
```

```python
df['seniorcitizen'] = df['seniorcitizen'].replace({
    0:'No',
    1:'Yes'
})
```

```python
df['tenure_group'] = pd.cut(
    df['tenure'],
    bins=[0,12,24,48,60,72],
    labels=[
        '0-12 Months',
        '13-24 Months',
        '25-48 Months',
        '49-60 Months',
        '61-72 Months'
    ]
)
```

```python
df.to_csv(
    "telecom_churn_cleaned.csv",
    index=False
)
```
